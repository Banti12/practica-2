        <h1> Practica 1 </h1>
        <h2> Primer paso instalar Nodejs</h2>
        <p> Una cosa que aprendí es que lo primero que tienes que hacer actualizar los paquetes y para eso necesitamos el comando
    <br> 
            <strong> sudo apt-get update </strong>
    <br> </p>
        <p> Para qué sirve el <strong> sudo </strong>. Es un comando que otorga privilegios de administrador a usuarios que no son administradores. </p>
        <p> Otra cosa que se requirió en la práctica es la instalación del NodeJs
        <br>
            <strong> sudo apt-get install nodejs </strong>
        <br>
        Nodejs es un código abierto de JavaScript que está diseñado para generar aplicaciones web de forma altamente optimizada. Es un sólo un lenguaje de programación del lado del cliente que se ejecuta en el navegador.
        </p>
        <br>
        <p>
            Se necesita instala el gestor de paquetes NPM de NodeJs
            <br>
                <strong> sudo apt-get install npm </strong>
            <br>
            NPM es un gestor de paquetes en lenguaje JavaScript, sirve para obtener cualquier librería con tan solo una sencilla línea de código, lo cual nos permitirá agregar dependencias de forma simple, distribuir paquetes y administrar eficazmente tanto los módulos como el proyecto a desarrollar en general.
        </p>
        <br>
        <p>
            También se necesita instalar el generador de aplicaciones Express
            <br>
                <strong>sudo npm install express-generator@4 –g</strong>
          Express.js es un framework para Node.js que sirve para ayudarnos a crear aplicaciones web en menos tiempo ya que nos proporciona funcionalidades como el  enrutamiento, opciones para gestionar sesiones, cookies, conectar por todas las funciones http,etc…  
            <br>
        </p>
        <br>
        <h2> Paso dos instalar PostgreSQL </h2>
        <br>
        <p>
            PostgreSQL es un sistema gestor de bases de datos relacionales, está orientado a objetos, es multiplataforma y open source.Para instalar este manejado de base de datos se necesita el siguiente comando <br>
            <strong>sudo apt install postgresql postgresql-contrib</strong> 
            <br>
            Una lista de comandos basicos de postgreSQL:
            <br>
            </p>
            <ul>
                <li>Ingresar al súperusuario postgres: <strong> sudo -i -u postgres</strong> </li>
                <li> Abrir PostgreSQL: <strong> psql </strong> </li>
                <li> Salir: <strong> \q </strong> </li>
                <li> Regresar al usuario: <strong> su - usuario </strong> </li>
                <li> Ingresar a PostgreSQL sin cambiar de usuario: <strong> sudo -u postgres psql </strong> </li>
                <li> Crear una base de datos como súperusuario: <strong> sudo -u postgres createdb test </strong> </li>
                <li> Crear una base desde PostgreSQL: <strong> createdb </strong> </li>
            </ul> 
    <br>
        <h2> Paso tres crearndo el proyecto </h2>|
        <p> Recomendación para crear el proyecto:</p>

            <li> Crear una carpeta para el proyecto <strong>sudo mkdir practicas cd practicas </strong> </li>

            <strong> sudo npm install pg-promise@5 –save </strong> <br>
            <strong> sudo npm install bluebird@3 –sabe </strong> <br>
            pg-promise no admite la cancelación de consultas, ya que es un truco para evitar el diseño incorrecto de la base de datos o la ejecución incorrecta de las consultas. <br>
            Bluebird es una biblioteca prometedora con todas las funciones que se centra en características y rendimiento innovadores.
